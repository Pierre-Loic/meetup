## üè¥‚Äç‚ò†Ô∏è Cr√©ation d'images de drapeaux √† partir de tableaux numpy

Voici une liste des quelques **fonctions numpy** utiles :

- numpy.full()


Fonctions proches : <a href="https://numpy.org/doc/stable/reference/generated/numpy.zeros.html" target="_blank">numpy.zeros()</a>, <a href="https://numpy.org/doc/stable/reference/generated/numpy.ones.html" target="_blank">numpy.ones()</a>

- numpy.concatenate()
- numpy.linspace()
- numpy.uint8

```python
# Dimensions de l'image
largeur = 1280 # pixels
hauteur = 720 # pixels
```

```python
# Image noire
image = np.zeros((hauteur, largeur, 3), dtype=np.uint8)

# Affichage de l'image
def show_image(image):
    plt.imshow(image)
    plt.axis('off')
    plt.show()
show_image(image)
```

```python
# Bandes de couleur pour le drapeau
bleu = (0, 85, 164)
blanc = (255, 255, 255)
rouge = (239, 65, 53)
bande_bleue = np.full((hauteur, largeur//3, 3), bleu, dtype=np.uint8)
bande_blanche = np.full((hauteur, largeur//3, 3), blanc, dtype=np.uint8)
bande_rouge = np.full((hauteur, largeur//3, 3), rouge, dtype=np.uint8)

# Concat√©nation des bandes pour former le drapeau
drapeau_francais = np.concatenate((bande_bleue, bande_blanche, bande_rouge), axis=1)

# Affichage du drapeau
plt.imshow(drapeau_francais)
plt.axis('off')
plt.show()
```

```python
# Bandes de couleur pour le drapeau
noir = (0, 0, 0)
jaune = (255, 233, 22)
rouge = (237, 41, 57)
bande_noir = np.full((hauteur, largeur//3, 3), noir, dtype=np.uint8)
bande_jaune = np.full((hauteur, largeur//3, 3), jaune, dtype=np.uint8)
bande_rouge = np.full((hauteur, largeur//3, 3), rouge, dtype=np.uint8)

# Concat√©nation des bandes pour former le drapeau
drapeau_belge = np.concatenate((bande_noir, bande_jaune, bande_rouge), axis=1)

# Affichage du drapeau
plt.imshow(drapeau_belge)
plt.axis('off')
plt.show()
```

```python
frames = []
for i in np.linspace(0, 1, 30): 
    intermediaire = (1 - i) * drapeau_francais + i * drapeau_belge
    frames.append(intermediaire.astype(np.uint8))

imageio.mimsave('transition_drapeau.gif', frames, duration=4, loop=0)
```

```python
path_to_gif = 'transition_drapeau.gif'
display(Image(filename=path_to_gif))
```
